```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tempvsAQI = pd.read_csv("TempvsAQI.csv")
df_tempvsPop = pd.read_csv("TempvsPopdensity.csv")
df_tempvsAQI.sort_values("AQI", inplace=True)
df_tempvsPop.sort_values("Population Density", inplace=True)
df_tempvsendType = pd.read_csv("FINGenomeVEndType.csv")
df_tempvsendType.sort_values("Genome Length", inplace=True)
df_tempvsGen = pd.read_csv("temperatethingy22.csv")
df_tempvsGen.sort_values("Genome Length", inplace=True)
```

```python
plot_tA = df_tempvsAQI.plot('Temperateness','AQI', kind = 'bar')
fig_tA = plot_tA.get_figure()
fig_tA.tight_layout()
fig_tA.savefig('Temperateness vs AQI.png')
```

```python
plot_pop = df_tempvsPop.plot('Temperateness','Population Density', kind = 'bar')
fig_pop = plot_pop.get_figure()
fig_pop.tight_layout()

fig_pop.savefig('Temperateness vs Population Density.png')
```

```python
plot_tE = df_tempvsendType.plot('End Type','Genome Length', kind = 'bar')
fig_tE = plot_tE.get_figure()
fig_tE.tight_layout()
fig_tE.savefig('End Type vs Genome Length.png')
```
