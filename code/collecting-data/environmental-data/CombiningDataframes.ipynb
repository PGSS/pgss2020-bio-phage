```python
import pandas as pd

#Read the factor CSVs into dataframe
PopDense = pd.read_csv('PopulationDensity.csv')
County = pd.read_csv('phages_county_all.csv')
WaterQuality = pd.read_csv('phwaterquality.csv')
Solar = pd.read_csv('NSRDB.csv')
AQI = pd.read_csv('df_AQI.csv')
Water2 = pd.read_csv("waterquality.csv")

#Drop pH values that are over 10
WaterQuality = WaterQuality[WaterQuality['pH'] < 10]
WaterQuality = pd.concat([WaterQuality, Water2])

#Merge needs to be 'outer' type to avoid dropping phages that are missing values for one of the factors, but have the values for other factors
Combined_Final = pd.merge(PopDense, County[["Phage Name","County"]], how='outer', on="Phage Name")
Combined_Final = pd.merge(Combined_Final, WaterQuality, how='outer', on=["GPSLatFormatted", "GPSLongFormatted"])


display(Combined_Final)
```

```python
Combined_Final = Combined_Final.drop_duplicates(subset='Phage Name') 
Combined_Final = Combined_Final.drop(columns='Unnamed: 0')

Combined_Final = pd.merge(Combined_Final, Solar, how='outer', on='Phage Name')


Combined_Final = Combined_Final.sort_values(by=['Phage Name'])
Combined_Final = pd.merge(Combined_Final, AQI, how='outer', left_on='Phage Name', right_on='PhageID')




Combined_Final = Combined_Final.reset_index(drop=True)
Combined_Final = Combined_Final.dropna(subset=['Phage Name'])
```

```python
#Drop extra columns
Combined_Final = Combined_Final.drop(columns=['lat_x', 'long_x', 'Unnamed: 0', 'PhageID', 'GPSLatFormatted', 'GPSLongFormatted'])
```

```python
Combined_Final = Combined_Final.rename(columns={'lat_y': 'Latitude', 'long_y': 'Longitude'})
Combined_Final = Combined_Final.reset_index(drop=True)
```

```python
Combined_Final = Combined_Final.rename(columns={'Zip': 'Zipcode'})
```

```python
#Drop duplicate phage entries
Combined_Final = Combined_Final.drop_duplicates(subset='Phage Name') 
```
