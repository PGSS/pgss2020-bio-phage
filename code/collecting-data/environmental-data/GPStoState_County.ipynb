```
def get_state(df, geolocator, latitude, longitude):
    location = geolocator.reverse((df[latitude], df[longitude]))
    try:
      state = location.raw['address']['state']
    except KeyError:
      state = None
      print("Unable to find state for", location.raw['address'])
    return state
```

```
def get_county(df, geolocator, latitude, longitude):
    location = geolocator.reverse((df[latitude], df[longitude]))
    try:
      county = location.raw['address']['county']
    except KeyError:
      county = None
      print("Unable to find county for", location.raw['address']['state'])
    return county
```

```
def get_zip(df, geolocator, latitude, longitude):
  location = geolocator.reverse((df[latitude], df[longitude]))
  try:
    zip = location.raw['address']['postcode']
  except KeyError:
      zip = None
      print("Unable to find zip for", location.raw['address']['county'])
  return zip
```

```
from geopy.geocoders import Nominatim
from geopy.point import Point
geolocator = Nominatim(user_agent="PGSS_CS_Bio")
```

```
df = pd.read_csv('https://raw.githubusercontent.com/amcguier/pgss2020-bio-phage/master/PhagesDB_Data_AllSequenced_NEW_FINAL.csv')
df
```

```
df = df[df["Found Country"] == "USA"]
```

```
df = df.dropna(subset=['GPSLatFormatted'])
```

```
df = df.dropna(subset=['GPSLongFormatted'])
```

```
df
```

```
df[['GPSLatFormatted', 'GPSLongFormatted', 'State', 'County']]
```

```
df.dtypes
```

```
df["State"] = df.apply(get_state, axis = 1, geolocator=geolocator, latitude="GPSLatFormatted", longitude="GPSLongFormatted")


```

```
df["County"] = df.apply(get_county, axis = 1, geolocator=geolocator, latitude="GPSLatFormatted", longitude="GPSLongFormatted")
```

```
df["Zip"] = df.apply(get_zip, axis = 1, geolocator=geolocator, latitude="GPSLatFormatted", longitude="GPSLongFormatted")
```
